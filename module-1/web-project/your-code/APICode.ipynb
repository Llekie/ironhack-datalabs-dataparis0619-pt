```python
import json
import requests
from pandas.io.json import json_normalize
```

```python
url='https://opendata.paris.fr/api/records/1.0/search/?dataset=que-faire-a-paris-&facet=category&facet=tags&facet=address_zipcode&facet=address_city&facet=pmr&facet=blind&facet=deaf&facet=access_type&facet=price_type'
```

```python
response = requests.get(url)
resp = response.content
results = response.json()
results


flattened_data = json_normalize(results)
flattened_data
```

```python
flattened_data1 = json_normalize(flattened_data.records[0])
flattened_data1
```

```python
flattened_data1.to_csv("QueFaireAParis.csv", sep = ";", index = False, encoding = "utf-8")
```
