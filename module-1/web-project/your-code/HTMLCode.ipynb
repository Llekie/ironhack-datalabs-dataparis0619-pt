```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
editeur = 'GALLIMARD'
collection ='Blanche'
titre = 'Une-amie-de-la-famille'
url = 'http://www.gallimard.fr/Catalogue/'+ editeur +'/'+ collection + '/'+ titre
```

```python
html = requests.get(url).content.decode("utf-8")
soup = BeautifulSoup(html, "lxml")
soup
```

```python
titrehtml = soup.find_all('div',{'class':"title"})
titreclean = [div.find_all('h1') for div in titrehtml]
lentitre = [y for x in titreclean for y in x if len(x) == 1]
titredef = [h1.text for h1 in lentitre]
vraititre = titredef[0]
print(vraititre)
```

```python
auteurhtml = soup.find_all('div',{'class':"author"})
auteurclean = [div.find_all('a') for div in auteurhtml]
lenauteur = [y for x in auteurclean for y in x if len(x) == 1]
auteurdef = [a.text for a in lenauteur]
auteur = auteurdef[0]
print(auteur)
```

```python
resumehtml = soup.find_all('div',{'class':"description expandable"})
resume = [div.text.replace('<br/>',' ').replace('\n','').replace('            ','') for div in resumehtml]
print(resume)
```

```python
dateparhtml = soup.find_all('div',{'class':"parution"})
dateparwip = [div.text.split(': ')for div in dateparhtml]
datepar = dateparwip[0][1]
print(datepar)
```

```python
EANhtml = soup.find('div',{'class':"info"})
EANextract = EANhtml.find('p').text.replace('  ','')
EANlist = re.findall('[0-9]{13}', EANextract)
EAN = EANlist[0]
print(EAN)
```

```python
print("Résumé de " + vraititre + " de " + auteur + " (" + EAN + "), paru en " + collection + " le " + datepar + " :")
print(resume[0])
```
