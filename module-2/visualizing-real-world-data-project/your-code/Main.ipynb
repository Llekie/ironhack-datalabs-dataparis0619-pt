```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
prem = pd.read_csv('..\cdsp_presi1965t1_circ.csv')
deux = pd.read_csv('..\cdsp_presi1965t2_circ.csv')
```

```python
#ajustement des types des contenus qualitatifs numériques
prem['Code département'] = prem['Code département'].astype(str)
prem['circonscription'] = prem['circonscription'].astype(str)
deux['Code département'] = deux['Code département'].astype(str)
deux['circonscription'] = deux['circonscription'].astype(str)
```

```python
prem['MITTERRAND'] = prem['MITTERRAND (CIR)']/prem['Inscrits']
prem['DE GAULLE'] = prem['DE GAULLE (UNR)']/prem['Inscrits']
prem['LECANUET'] = prem['LECANUET (MRP)']/prem['Inscrits']
prem['MARCILHACY'] = prem['MARCILHACY (DVD)']/prem['Inscrits']
prem['TIXIER-VIGNANCOUR'] = prem['TIXIER-VIGNANCOUR (EXD)']/prem['Inscrits']
prem['BARBU'] = prem['BARBU (DIV)']/prem['Inscrits']
prem['Blancs et nuls%'] = prem['Blancs et nuls']/prem['Inscrits']
prem['Exprimés%'] = prem['Exprimés']/prem['Inscrits']
prem['circonunique'] = prem.département +" - "+ prem.circonscription

deux['MITTERRAND'] = deux['MITTERRAND (CIR)']/deux['Inscrits']
deux['DE GAULLE'] = deux['DE GAULLE (UNR)']/deux['Inscrits']
deux['Blancs et nuls%'] = deux['Blancs et nuls']/deux['Inscrits']
deux['Exprimés%'] = deux['Exprimés']/deux['Inscrits']
deux['circonunique'] = deux.département +" - "+ deux.circonscription
```

```python
fig, [scat,scat1,scat2,scat3] = plt.subplots(figsize = (15,6),nrows = 1,ncols = 4)
#Votes pour De Gaulle
scat.scatter(prem.Inscrits, prem['Exprimés%'], alpha=0.6, c='r')
scat.scatter(deux.Inscrits, deux['Exprimés%'], alpha=0.6)
scat.set_xlabel("nb de votes exprimés par circonscription")
scat.set_ylabel("nb d'inscrits par circonscription")
scat.set_title("Comparaison des 2 tours de \n l'élection présidentielle de 65 : \n votes exprimés vs inscrits")
scat.legend(['1er tour', '2nd tour'])
scat.set_ylim(0,1)
#Votes pour De Gaulle
scat1.scatter(prem.Inscrits, prem['DE GAULLE'], alpha=0.6, c='r')
scat1.scatter(deux.Inscrits, deux['DE GAULLE'], alpha=0.6)
scat1.set_xlabel("nb de votes pour De Gaulle par circonscription")
scat1.set_ylabel("nb d'inscrits par circonscription")
scat1.set_title("Comparaison des 2 tours de \n l'élection présidentielle de 65 : \n votes pour De Gaulle par circonscription")
scat1.legend(['1er tour', '2nd tour'])
scat1.set_ylim(0,1)
#Votes pour Mitterrand
scat2.scatter(prem.Inscrits, prem['MITTERRAND'], alpha=0.6, c='r')
scat2.scatter(deux.Inscrits, deux['MITTERRAND'], alpha=0.6)
scat2.set_xlabel("nb de votes pour Mitterrand par circonscription")
scat2.set_ylabel("nb d'inscrits par circonscription")
scat2.set_title("Comparaison des 2 tours de \n l'élection présidentielle de 1965 : \n votes pour Mitterrand par circonscription")
scat2.legend(['1er tour', '2nd tour'])
scat2.set_ylim(0,1)
#Votes blancs et nuls
scat3.scatter(prem.Inscrits, prem['Blancs et nuls%'], alpha=0.6, c='r')
scat3.scatter(deux.Inscrits, deux['Blancs et nuls%'], alpha=0.6)
scat3.set_xlabel("nb de votes pour Mitterrand par circonscription")
scat3.set_ylabel("nb d'inscrits par circonscription")
scat3.set_title("Comparaison des 2 tours de \n l'élection présidentielle de 1965 : \n votes blancs et nuls par circonscription \n (attention échelle !)")
scat3.legend(['1er tour', '2nd tour'])
scat3.set_ylim(0,0.1)
plt.show()
```

```python
#Entre le 1er et le 2nd tour, on note une très faible érosion du nombre de votants.
#les 2 candidats ont bien sur chacun gagné des voix.
#La progression de Mitterrand a été plus importante que celle de De Gaulle.
#Les votes blancs et nuls, bien très minoritaires, ne peuvent pas être représentés à la même échelle.
#Mais l'évolution de ce vote est bien plus marquée que pour les 2 candidats (x2 au moins).
```

```python
fig, [hist1,hist2,hist3] = plt.subplots(figsize = (15,6),nrows = 1,ncols = 3)
departmt = prem.groupby('circonunique').sum()
#Hist De Gaulle
hist1.hist(departmt['DE GAULLE'], bins=40)
hist1.set_xlim(0,1)
hist1.set_ylim(0,50)
hist1.set_title("Représentation du poids du vote DE GAULLE \n par circonscription \n (1er tour de l'élection de 1965)")
hist1.set_xlabel("Poids du vote pour De Gaulle par circons.")
#Hist De Mitterrand
hist2.hist(departmt['MITTERRAND'], bins=40)
hist2.set_xlim(0,1)
hist2.set_ylim(0,50)
hist2.set_title("Représentation du poids du vote MITTERRAND \n par circonscription \n (1er tour de l'élection de 1965)")
hist2.set_xlabel("Poids du vote pour Mitterrand par circons.")
#Hist Lecanuet
hist3.hist(departmt['LECANUET'], bins=40)
hist3.set_xlim(0,1)
hist3.set_ylim(0,50)
hist3.set_title("Représentation du poids du vote LECANUET \n par circonscription \n (1er tour de l'élection de 1965)")
hist3.set_xlabel("Poids du vote pour Lecanuet par circons.")
plt.show()
```

```python
#Ces histogrammes montrent la répartition du poids de vote pour les 3 principaux candidats, en nombre de circonscriptions.
#Ils permettent ici de voir quel est le gagnat : l'histogramme concernant De Gaulle est centré vers les poids les plus importants
#Centre visuel de la répartition pour De Gaulle : 40% / Pour Mitterrand : 25% / Pour Lecanuet : 18%.
```

```python
graph2 = prem.boxplot(column=['DE GAULLE','MITTERRAND','LECANUET','TIXIER-VIGNANCOUR','MARCILHACY','BARBU'],figsize=(14, 8))
graph2.set_title('1er tour des présidentielles 1965 : Poids du vote en faveur des différents candidats par circonscription')
graph2.set_xlabel('candidats')
graph2.set_ylabel('Poids du vote en faveur du candidat sur la circonscription')
plt.show()
```

```python
#Ces boites à moustache sont une autre façon de visualiser la répartition du vote pour les différents candidats.
#Il permet d'identifier plus précisément le centre de la distribution.
# A noter :
#De Gaulle : vote globalement très concentré. Absence de circonscription avec un score exceptionnellement bas, et inversement, quelques bastions acquis à sa cause.
# Mitterrand : le vote est bcp plus aléatoire que pour De Gaulle (répartition bcp plus hétérogène)
#Tixier-Vignancour : vote caractéristique de l'extrême droite : scores plutôt faibles, avec toute une série de circonscriptions "outsiders" acquises à la cause
#Marcilhacy : on identifie parfaitement les circonscriptions autour de chez lui, les points isolés qui lui ont permis de récolter quelques voix !
```

```python
prov = prem.groupby('département').sum().sort_values('Inscrits', ascending = False)#[0:20]
prov['MITTERRAND'] = prov['MITTERRAND (CIR)']/prov['Votants']
prov['DE GAULLE'] = prov['DE GAULLE (UNR)']/prov['Votants']
prov['LECANUET'] = prov['LECANUET (MRP)']/prov['Votants']
prov['MARCILHACY'] = prov['MARCILHACY (DVD)']/prov['Votants']
prov['TIXIER-VIGNANCOUR'] = prov['TIXIER-VIGNANCOUR (EXD)']/prov['Votants']
prov['BARBU'] = prov['BARBU (DIV)']/prov['Votants']
prov['Blancs et nuls%'] = prov['Blancs et nuls']/prov['Votants']
prov['Exprimés%'] = prov['Exprimés']/prov['Votants']

graphcum = prov[['DE GAULLE','MITTERRAND','LECANUET','TIXIER-VIGNANCOUR','MARCILHACY','BARBU','Blancs et nuls%']]
graphcum2 = prov[['DE GAULLE (UNR)','MITTERRAND (CIR)','LECANUET (MRP)','TIXIER-VIGNANCOUR (EXD)','MARCILHACY (DVD)','BARBU (DIV)','Blancs et nuls']]
```

```python
ind = np.arange(len(graphcum['DE GAULLE']))
p1 = plt.bar(ind, graphcum['DE GAULLE'].values, width = 0.4)
p2 = plt.bar(ind, graphcum['MITTERRAND'].values, bottom = graphcum['DE GAULLE'].values, width = 0.4)
p3 = plt.bar(ind, graphcum['LECANUET'].values, bottom = (graphcum['DE GAULLE'].values + graphcum['MITTERRAND'].values), width = 0.4)
p4 = plt.bar(ind, graphcum['TIXIER-VIGNANCOUR'].values, bottom = graphcum['DE GAULLE'].values + graphcum['MITTERRAND'].values + graphcum['LECANUET'].values, width = 0.4)
p5 = plt.bar(ind, graphcum['MARCILHACY'].values, bottom = graphcum['DE GAULLE'].values + graphcum['MITTERRAND'].values + graphcum['LECANUET'].values + graphcum['TIXIER-VIGNANCOUR'].values, width = 0.4)
p6 = plt.bar(ind, graphcum['BARBU'].values, bottom = graphcum['DE GAULLE'].values + graphcum['MITTERRAND'].values + graphcum['LECANUET'].values + graphcum['TIXIER-VIGNANCOUR'].values+ graphcum['MARCILHACY'].values, width = 0.4)
p7 = plt.bar(ind, graphcum['Blancs et nuls%'].values, bottom = graphcum['DE GAULLE'].values + graphcum['MITTERRAND'].values + graphcum['LECANUET'].values + graphcum['TIXIER-VIGNANCOUR'].values+ graphcum['MARCILHACY'].values+ graphcum['BARBU'].values, width = 0.4)
plt.xticks(ind, graphcum.index, rotation = 'vertical')
plt.rcParams["figure.figsize"]=(5,5)
plt.ylabel('poids du vote pour chaque candidat')
plt.xlabel('départements')
plt.title('Poids du vote pour chaque candidat sur le total des votes exprimés par département')
plt.legend((p1[0],p2[0],p3[0],p4[0],p5[0],p6[0],p7[0]), ('DE GAULLE','MITTERRAND','LECANUET','TIXIER-VIGNANCOUR','MARCILHACY','BARBU','Blancs et nuls'), fontsize= 7)
plt.show()
```

```python
ind = np.arange(len(graphcum2['DE GAULLE (UNR)']))
p1 = plt.bar(ind, graphcum2['DE GAULLE (UNR)'].values, width = 0.4)
p2 = plt.bar(ind, graphcum2['MITTERRAND (CIR)'].values, bottom = graphcum2['DE GAULLE (UNR)'].values, width = 0.4)
p3 = plt.bar(ind, graphcum2['LECANUET (MRP)'].values, bottom = (graphcum2['DE GAULLE (UNR)'].values + graphcum2['MITTERRAND (CIR)'].values), width = 0.4)
p4 = plt.bar(ind, graphcum2['TIXIER-VIGNANCOUR (EXD)'].values, bottom = graphcum2['DE GAULLE (UNR)'].values + graphcum2['MITTERRAND (CIR)'].values + graphcum2['LECANUET (MRP)'].values, width = 0.4)
p5 = plt.bar(ind, graphcum2['MARCILHACY (DVD)'].values, bottom = graphcum2['DE GAULLE (UNR)'].values + graphcum2['MITTERRAND (CIR)'].values + graphcum2['LECANUET (MRP)'].values + graphcum2['TIXIER-VIGNANCOUR (EXD)'].values, width = 0.4)
p6 = plt.bar(ind, graphcum2['BARBU (DIV)'].values, bottom = graphcum2['DE GAULLE (UNR)'].values + graphcum2['MITTERRAND (CIR)'].values + graphcum2['LECANUET (MRP)'].values + graphcum2['TIXIER-VIGNANCOUR (EXD)'].values+ graphcum2['MARCILHACY (DVD)'].values, width = 0.4)
p6 = plt.bar(ind, graphcum2['Blancs et nuls'].values, bottom = graphcum2['DE GAULLE (UNR)'].values + graphcum2['MITTERRAND (CIR)'].values + graphcum2['LECANUET (MRP)'].values + graphcum2['TIXIER-VIGNANCOUR (EXD)'].values+ graphcum2['MARCILHACY (DVD)']+ graphcum2['BARBU (DIV)'].values, width = 0.4)
plt.xticks(ind, graphcum2.index, rotation = 'vertical')
plt.rcParams["figure.figsize"]=(10,10)
plt.ylabel('vote en faveur de chaque candidat')
plt.xlabel("départements (classés par nombre dégressif d'inscrits sur les listes électorales)")
plt.title('Nombre de votes pour chaque candidat par département')
plt.legend((p1[0],p2[0],p3[0],p4[0],p5[0],p6[0],p7[0]), ('DE GAULLE','MITTERRAND','LECANUET','TIXIER-VIGNANCOUR','MARCILHACY','BARBU','Blancs et nuls'), loc = 'best')
plt.show()
```

```python
#Le premier graphique en barre permet de constater une relative homogénéité entre le vote des différents départements.
#On voit apparaître malgré tout quelques régions très attachées à De Gaulle (l'Alsace) ;
#les départements concernés sont ceux qui ont le moins voté pour Mitterrand.
#On identifie également clairement le poids pris par Marcilhacy sur sa région, la Charente.

#Le second graphique permet quant à lui d'identifier le poids particulièrement important des 3 premiers départements,
#complètement disproportionné par rapport aux autres départements.
#On voit également un "trou" au niveau de la Corse :
#le graphique est classé par dégressif d'inscrits, les valeurs renseignées correspondent aux votants.
#ce trou correspond donc à une forte proportion d'abstentionnistes.
```

```python
def barchart(col):
    s = prem.groupby('département').sum()
    s = s[col].sort_values(ascending=False)[:10]
    s = s.sort_values()
    return s.plot.barh()

barchart('BARBU')
```
