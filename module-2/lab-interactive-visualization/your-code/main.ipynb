## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
April11 = data[(data.Country != "United Kingdom") & data.InvoiceDate.astype(str).str.contains("2011-04")]
April11.groupby('Country', as_index = False).sum()
April11[['Country','Revenue','Quantity']].iplot(kind='bar', x = 'Country', xTitle='Country', yTitle='Revenue & Quantity', 
           title='April 2011 : Quantites and revenues by Country (excluding UK)')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
data['Date'] = pd.to_datetime(data['InvoiceDate']).dt.strftime('%Y.%m.%d')
France = data[(data.Country == "France") & data.InvoiceDate.between('2011-01-01','2011-05-31')].sort_values('Date')
France = France.groupby('Date', as_index = False).sum()
France[['Date','Revenue','Quantity']].iplot(kind='line', x = 'Date', xTitle='Date', yTitle='Revenue & Quantity', 
           title='janvier à mai 2011 : Quantites et revenus en France')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
PartyBunting = data[(data.Description == "PARTY BUNTING")].sort_values('Quantity')
PartyBunting = PartyBunting.groupby('Country', as_index = False).mean()
PartyBunting[['Country','UnitPrice','Quantity']].iplot(x = 'Quantity', xTitle='Quantity', y = 'UnitPrice', yTitle='UnitPrice', 
           categories='Country' , title='moyennes des quantités et du prix unitaire par pays')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
Pays = ['EIRE', 'Germany', 'France', 'Netherlands']
QtyPerInvoice = data[data.Country.isin(Pays)].pivot_table(values='Quantity', columns='Country', index="InvoiceNo", aggfunc='sum')
QtyPerInvoice.iplot(kind='hist', histnorm='percent', xTitle='Quantité', subplots=True,
yTitle='Nb de commandes', title='Distribution de la valeur commandée par pays')
```

## 5. Create an interactive side-by-side bar chart showing the revenue by country listed below (bars) for each of the products listed below.

```python
product_list = ['JUMBO BAG RED RETROSPOT', 
                'CREAM HANGING HEART T-LIGHT HOLDER',
                'REGENCY CAKESTAND 3 TIER']

country_list = ['EIRE', 'Germany', 'France', 'Netherlands']
```

```python
RevByCountry = data[(data.Description.isin(product_list)) & (data.Country.isin(country_list))]
RevByCountry = RevByCountry.pivot_table(values='Revenue', columns='Description', index="Country", aggfunc='sum', )
RevByCountry['Country'] = RevByCountry.index
RevByCountry.iplot(kind='bar', x = 'Country', xTitle='Country', yTitle='Revenue', 
           title='Revenues Description and Country')
```
