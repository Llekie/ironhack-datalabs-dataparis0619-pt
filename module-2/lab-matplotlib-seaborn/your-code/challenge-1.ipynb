#### Define data.

#### There are 2 ways of doing this. Do in both ways.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.subplot(1, 2, 2)
plt.plot(x,z)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1,2, figsize=[12, 4])
ax1.plot(x,y, color = 'r', linestyle = '--')
ax2.plot(x,z, color = 'g', linestyle = ':')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1,2, figsize=[12, 4])
ax1.plot(x,y, color = 'r', linestyle = '--')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,z, color = 'g', linestyle = ':')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1,2, figsize=[12, 4])

#plot 1 log
y = x**2
ax1.plot(x,y, color = 'r', linestyle = '--')
y=np.exp(x)
ax1.plot(x,y, color = 'g', linestyle = ':')
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_title('log')

#plot 2 linear
y = x**2
ax2.plot(x,y, color = 'r', linestyle = '--')
y=np.exp(x)
ax2.plot(x,y, color = 'g', linestyle = ':')
ax2.set_yscale('linear')
ax2.set_xscale('linear')
ax2.set_title('linear')
```

#### In the vehicles data set that you have downloaded, conduct exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. Please provide at least one inssight from the plot that you have drawn

Scatter Plot

```python
# your code here
data = pd.read_csv('vehicles.csv')
data.head()
```

```python
y=data['Combined MPG']
x=data['CO2 Emission Grams/Mile']
plt.scatter(x,y, color = 'b', s = 1.5, alpha=0.2)
plt.xlabel('CO2 Emission Grams/Mile')
plt.ylabel('Combined MPG')
plt.title("Corrélation entre le MPG et l'émission de CO2")
plt.show()
```

```python
# il y a une forte corrélation négative non-linéaire
```

Box Whisker Plot

```python
# your code here
Regular=data[data['Fuel Type'] == "Regular"]['Fuel Cost/Year']
Premium=data[data['Fuel Type'] == "Premium"]['Fuel Cost/Year']
E85=data[data['Fuel Type'] == "Gasoline or E85"]['Fuel Cost/Year']
Diesel=data[data['Fuel Type'] == "Diesel"]['Fuel Cost/Year']
plt.boxplot([Regular,Premium,E85,Diesel], labels=["Regular","Premium","E85","Diesel"])
plt.title('prix du carburant par an par type de carburant')
plt.show()
```

```python
#La distribution du prix du carburant dépend du type de carburant.
#La distribution la plus resserrée correspond au prix du Premium, avec toutefois beaucoup de valeurs atypiques extrêmes
```

Histogram

```python
# your code here
y=data['Fuel Cost/Year']
plt.hist(y, bins=50)
plt.xlabel('nb lignes échantillon')
plt.ylabel('Years')
plt.title("nb de valeurs disponibles dans l'échantillon par an")
plt.show()
```

```python
#La distribution par an de l'échantillon ne suit pas la loi normale
```

Bar Chart

```python
# your code here
y=data['CO2 Emission Grams/Mile']
x=data['Fuel Type']
plt.bar(x,y, color = 'g')
plt.xlabel('Fuel Type')
plt.xticks(rotation = 'vertical')
plt.ylabel('CO2 Emission Grams/Mile')
plt.title('Fuel Type vs CO2 Emission Grams/Mile')
plt.show()
```

```python
# C'est le Fuel Premium qui rejette le plus de CO2.
#L'électricité est le type de carburant qui en rejette le moins.
```
